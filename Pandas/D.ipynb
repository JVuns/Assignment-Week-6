```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
filepath = "D:/Ran/activity.csv"
pd.set_option("display.max_rows", None, "display.max_columns", None)
filepath = "D:/Ran/activity.csv"
```

```python
df = pd.read_csv(filepath)

days = []
steps = []
isweekdayid = []
isweekday = []

df['whatweek'] = pd.to_datetime(df['date']).dt.dayofweek 
df["whatweek"] = (df["whatweek"] < 5).astype(float)
# print(df)

for x in df['date'].unique():
    days.append(x)
    steps.append(df.loc[df['date']==x, 'steps'].sum())
    isweekdayid.append(df.loc[df['date']==x, 'whatweek'].sum())
print(isweekday)
for x in isweekdayid:
    if x == 288.0:
        isweekday.append('weekday')
    else:
        isweekday.append('weekend')

df = pd.DataFrame({'Days':days, 'Steps':steps, 'Isweekday':isweekday})
print(df)
```

```python
df.plot(x ='Days', kind = 'line')
plt.title('average number of steps taken')
plt.xticks(np.linspace(0,61,5))
plt.ylabel('Steps')
plt.show()
```
